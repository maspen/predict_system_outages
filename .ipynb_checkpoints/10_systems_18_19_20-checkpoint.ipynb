```python
def get_outage_non_outage_time_span_array(df):
    first_prob_start = df['problem_start_datetime'].min()
    # print(first_prob_start)
    last_prob_end = df['problem_end_datetime'].max()
    # print(last_prob_end)
    delta_mins = (last_prob_end - first_prob_start) / pd.Timedelta(minutes=1)
    
    time_span_array = np.full(int(delta_mins), -1, dtype=int)
    for idx, row in df.iterrows():
        row_outage_start = row.problem_start_datetime - first_prob_start
        row_outage_start_mins = int(row_outage_start / pd.Timedelta(minutes=1))
        row_outage_ends = row.problem_end_datetime - first_prob_start
        row_outage_end_mins = int(row_outage_ends / pd.Timedelta(minutes=1))
        time_span_array[row_outage_start_mins:row_outage_end_mins] = idx
        
         # print(f'idx: {idx} s: {row.problem_start_datetime} e: {row.problem_end_datetime} s_m {row_outage_start_mins} e_m {row_outage_end_mins}')

    return time_span_array

def create_non_outage_row(row, problem_start_datetime, problem_end_datetime):
    row_copy = row.copy()
    down_time_mins = int((problem_end_datetime - problem_start_datetime)/pd.Timedelta(minutes=1))
    row_copy['problem_start_datetime'] = problem_start_datetime.to_datetime64()
    row_copy['problem_end_datetime'] = problem_end_datetime.to_datetime64()
    row_copy['down_time_mins'] = down_time_mins
    row_copy['is_outage'] = 'no'
    return row_copy

def add_row_to_df(df, row):
    df.loc[len(df),:] = row
    
def add_non_outage_rows(df_outages, arr):
    df = df_outages.copy()
    
    first_outage_start = df['problem_start_datetime'].min()

    j = 0
    end = len(arr)
    while j < end:
        if arr[j] == -1:
            row_idx = arr[j - 1]
            start_idx = j
            while arr[j] == -1:
                if (j+1) < end:
                    j += 1
                else:
                    break
            end_idx = j - 1
            
            start_timestamp = first_outage_start + pd.Timedelta(minutes=start_idx)
            end_timestamp = start_timestamp + pd.Timedelta(minutes=end_idx)
            non_outage_row = create_non_outage_row(df.iloc[row_idx], start_timestamp, end_timestamp)
            df.loc[len(df),:] = non_outage_row
        else:
            j += 1
            
    df.sort_values(by=['problem_start_datetime'], ascending=True, inplace=True)
    df.reset_index(inplace=True)
            
    print('done')
    return df    
```

```python
# https://scikit-learn.org/stable/auto_examples/impute/plot_iterative_imputer_variants_comparison.html

from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.kernel_approximation import Nystroem
from sklearn.linear_model import BayesianRidge, Ridge
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor

N_SPLITS = 4

rng = np.random.RandomState(42)

y_full = df_5.iloc[:,-1:]
X_full = df_5.loc[:, ~df_5.columns.isin(['is_outage'])]

n_samples, n_features = X_full.shape

# Estimate the score on the entire dataset, with no missing values
# br_estimator = BayesianRidge()
# score_full_data = pd.DataFrame(
#     cross_val_score(
#         br_estimator, X_full, y_full, scoring="neg_mean_squared_error", cv=N_SPLITS
#     ),
#     columns=["Full Data"],
# )

# Add a single missing value to each row
X_missing = X_full.copy()
y_missing = y_full
# missing_samples = np.arange(n_samples)
# missing_features = rng.choice(n_features, n_samples, replace=True)
# X_missing[missing_samples, missing_features] = np.nan

# Estimate the score after imputation (mean and median strategies)
score_simple_imputer = pd.DataFrame()
for strategy in ("mean", "median"):
    estimator = make_pipeline(
        SimpleImputer(missing_values=np.nan, strategy=strategy), br_estimator
    )
    score_simple_imputer[strategy] = cross_val_score(
        estimator, X_missing, y_missing, scoring="neg_mean_squared_error", cv=N_SPLITS
    )

# Estimate the score after iterative imputation of the missing values
# with different estimators
estimators = [
    BayesianRidge(),
    RandomForestRegressor(
        # We tuned the hyperparameters of the RandomForestRegressor to get a good
        # enough predictive performance for a restricted execution time.
        n_estimators=4,
        max_depth=10,
        bootstrap=True,
        max_samples=0.5,
        n_jobs=2,
        random_state=42,
    ),
    make_pipeline(
        Nystroem(kernel="polynomial", degree=5, random_state=42), Ridge(alpha=1e3)
    ),
    KNeighborsRegressor(n_neighbors=15),
]
score_iterative_imputer = pd.DataFrame()
# iterative imputer is sensible to the tolerance and
# dependent on the estimator used internally.
# we tuned the tolerance to keep this example run with limited computational
# resources while not changing the results too much compared to keeping the
# stricter default value for the tolerance parameter.
tolerances = (1e-3, 1e-1, 1e-1, 1e-2)
for impute_estimator, tol in zip(estimators, tolerances):
    estimator = make_pipeline(
        IterativeImputer(
            random_state=42, estimator=impute_estimator, max_iter=25, tol=tol
        ),
        br_estimator,
    )
    score_iterative_imputer[impute_estimator.__class__.__name__] = cross_val_score(
        estimator, X_missing, y_missing, scoring="neg_mean_squared_error", cv=N_SPLITS
    )

scores = pd.concat(
    [score_full_data, score_simple_imputer, score_iterative_imputer],
    keys=["Original", "SimpleImputer", "IterativeImputer"],
    axis=1,
)

# plot california housing results
fig, ax = plt.subplots(figsize=(13, 6))
means = -scores.mean()
errors = scores.std()
means.plot.barh(xerr=errors, ax=ax)
ax.set_title("System 18,19,20 Imputer comparison")
ax.set_xlabel("MSE (smaller is better)")
ax.set_yticks(np.arange(means.shape[0]))
ax.set_yticklabels([" w/ ".join(label) for label in means.index.tolist()])
plt.tight_layout(pad=1)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import ExtraTreesRegressor

X = df_5.loc[:, ~df_5.columns.isin(['is_outage'])]
y = df_5.iloc[:,-1:]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

params = {
    'i__missing_values':[np.nan],
    'i__tol':[1e-3, 1e-2, 1e-1],
    'm__criterion':['squared_error','friedman_mse'],
    'm__max_depth':[4,8],
    'm__max_features':['sqr','log2']
}

pipelin_0 = Pipeline([
    ('s', StandardScaler()),
    ('i', IterativeImputer(random_state=42)),
    ('m', ExtraTreesRegressor(random_state=42, n_jobs=-1))
])

imputer_grid = GridSearchCV(pipelin_0, params, scoring='neg_mean_squared_error', cv=5).fit(X_train, y_train.values.ravel())
```

```python
imputer_grid_score = imputer_grid.score(X_test, y_test)
imputer_grid_best_params = imputer_grid.best_params_
imputer_grid_best_estimator = imputer_grid.best_estimator_

print(f'score {imputer_grid_score}')
print(f'best params {imputer_grid_best_params}')
print(f'best estimator {imputer_grid_best_estimator}')
```
